promedio, nota más alta y nota más baja




```python
def calcular_promedio(notas):
    if len(notas) == 0:
        return 0
    return sum(notas) / len(notas)

def nota_mas_alta(notas):
    if len(notas) == 0:
        return None
    return max(notas)

def nota_mas_baja(notas):
    if len(notas) == 0:
        return None
    return min(notas)

notas = [3.5, 4.2, 2.8, 5.0, 3.9]

print("Promedio:", calcular_promedio(notas))
print("Nota más alta:", nota_mas_alta(notas))
print("Nota más baja:", nota_mas_baja(notas))
```

agenda simple

```python
def agregar_contacto(agenda, nombre, telefono):
    agenda[nombre] = telefono

def buscar_contacto(agenda, nombre):
    if nombre in agenda:
        return agenda[nombre]
    return "Contacto no encontrado"

def eliminar_contacto(agenda, nombre):
    if nombre in agenda:
        del agenda[nombre]
        return "Contacto eliminado"
    return "Contacto no encontrado"

agenda = {}

agregar_contacto(agenda, "Ana", "3001234567")
agregar_contacto(agenda, "Luis", "3119876543")

print("Buscar Ana:", buscar_contacto(agenda, "Ana"))
print(eliminar_contacto(agenda, "Luis"))
print("Agenda actual:", agenda)
```

estudiantes aprobados

```python
def obtener_aprobados(estudiantes):
    aprobados = []
    for nombre, nota in estudiantes:
        if nota >= 3.0:
            aprobados.append((nombre, nota))
    return aprobados

estudiantes = [
    ("Ana", 3.5),
    ("Luis", 2.7),
    ("Carlos", 4.0),
    ("Marta", 2.9),
    ("Sofía", 5.0)
]

print("Estudiantes aprobados:")
print(obtener_aprobados(estudiantes))
```

inventario y valor total

```python
def calcular_valor_total_inventario(inventario):
    total = 0
    for nombre, datos in inventario.items():
        total += datos["precio"] * datos["cantidad"]
    return total

inventario = {
    "Lapiz": {"precio": 1500, "cantidad": 10},
    "Cuaderno": {"precio": 5000, "cantidad": 5},
    "Borrador": {"precio": 1200, "cantidad": 8}
}

print("Valor total del inventario:", calcular_valor_total_inventario(inventario))
```

frecuencia de palabras

```python
def contar_frecuencia_palabras(lista_palabras):
    frecuencias = {}
    for palabra in lista_palabras:
        if palabra in frecuencias:
            frecuencias[palabra] += 1
        else:
            frecuencias[palabra] = 1
    return frecuencias

palabras = ["sol", "luna", "sol", "estrella", "luna", "sol"]

print("Frecuencia de palabras:")
print(contar_frecuencia_palabras(palabras))
```

ciudad más caliente y más fría

```python
def ciudad_mas_caliente_y_fria(temperaturas_ciudades):
    ciudad_caliente = None
    temp_max = float("-inf")

    ciudad_fria = None
    temp_min = float("inf")

    for ciudad, temperaturas in temperaturas_ciudades.items():
        maxima = max(temperaturas)
        minima = min(temperaturas)

        if maxima > temp_max:
            temp_max = maxima
            ciudad_caliente = ciudad

        if minima < temp_min:
            temp_min = minima
            ciudad_fria = ciudad

    return (ciudad_caliente, temp_max), (ciudad_fria, temp_min)

temperaturas = {
    "Bogotá": [18, 17, 19, 20, 16, 18, 17],
    "Medellín": [28, 29, 30, 31, 27, 28, 29],
    "Cali": [30, 32, 31, 33, 29, 30, 31]
}

mas_caliente, mas_fria = ciudad_mas_caliente_y_fria(temperaturas)

print("Ciudad más caliente:", mas_caliente)
print("Ciudad más fría:", mas_fria)
```

convertir notas numéricas a letras

```python
rangos_notas = (
    (4.5, 5.0, "A"),
    (4.0, 4.49, "B"),
    (3.0, 3.99, "C"),
    (2.0, 2.99, "D"),
    (0.0, 1.99, "F")
)

def convertir_nota_a_letra(nota):
    for minimo, maximo, letra in rangos_notas:
        if minimo <= nota <= maximo:
            return letra
    return "Nota inválida"

def generar_reporte_estudiantes(estudiantes):
    reporte = []
    for nombre, nota in estudiantes:
        letra = convertir_nota_a_letra(nota)
        reporte.append((nombre, nota, letra))
    return reporte

estudiantes = [
    ("Ana", 4.8),
    ("Luis", 4.2),
    ("Carlos", 3.5),
    ("Marta", 2.3),
    ("Pedro", 1.8)
]

print("Reporte de estudiantes:")
print(generar_reporte_estudiantes(estudiantes))
```

carrito de compras

```python
def agregar_al_carrito(carrito, nombre, precio, cantidad):
    carrito.append({
        "nombre": nombre,
        "precio": precio,
        "cantidad": cantidad
    })

def calcular_subtotal(carrito):
    subtotal = 0
    for producto in carrito:
        subtotal += producto["precio"] * producto["cantidad"]
    return subtotal

def aplicar_descuento(total, porcentaje_descuento):
    return total - (total * porcentaje_descuento / 100)

carrito = []

agregar_al_carrito(carrito, "Camisa", 50000, 2)
agregar_al_carrito(carrito, "Pantalón", 80000, 1)
agregar_al_carrito(carrito, "Zapatos", 120000, 1)

subtotal = calcular_subtotal(carrito)
total_final = aplicar_descuento(subtotal, 10)

print("Subtotal:", subtotal)
print("Total con descuento:", total_final)
```

agrupar productos por categoría

```python
def agrupar_productos_por_categoria(productos):
    categorias = {}
    for producto, categoria in productos:
        if categoria not in categorias:
            categorias[categoria] = []
        categorias[categoria].append(producto)
    return categorias

productos = [
    ("Manzana", "Frutas"),
    ("Pera", "Frutas"),
    ("Zanahoria", "Verduras"),
    ("Lechuga", "Verduras"),
    ("Leche", "Lácteos")
]

print("Productos agrupados por categoría:")
print(agrupar_productos_por_categoria(productos))
```

sistema de votos

```python
def registrar_votos(lista_votos, candidatos_validos):
    conteo = {}
    votos_invalidos = 0

    for candidato in candidatos_validos:
        conteo[candidato] = 0

    for voto in lista_votos:
        if voto in candidatos_validos:
            conteo[voto] += 1
        else:
            votos_invalidos += 1

    total_validos = sum(conteo.values())

    if total_validos > 0:
        ganador = max(conteo, key=conteo.get)
        porcentaje = (conteo[ganador] / total_validos) * 100
    else:
        ganador = None
        porcentaje = 0

    return conteo, votos_invalidos, ganador, porcentaje

candidatos = ["Ana", "Luis", "Carlos"]
votos = ["Ana", "Luis", "Ana", "Carlos", "Pedro", "Ana", "Luis", "Juan", "Carlos", "Ana"]

resultado = registrar_votos(votos, candidatos)

print("Conteo de votos:", resultado[0])
print("Votos inválidos:", resultado[1])
print("Ganador:", resultado[2])
print("Porcentaje del ganador:", resultado[3])
```
